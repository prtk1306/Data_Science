```python
#!/bin/python3

import math
import os
import random
import re
import sys

arr = []

for _ in range(6):
    arr.append(list(map(int, input().rstrip().split())))



maxx = -999999999999999

for i in range(0, 6-2):
    for j in range(1,5):
        count = []
        count.append(arr[i][j-1: j+2])
        count.append(arr[i+1][j])
        count.append(arr[i+2][j-1: j+2])
        
        total = 0
            
        for k in count:

            if type(k) is list:
                total += sum(k)
            else:
                total += k
        
        if maxx < total:
            maxx = total
            
print(maxx)
```
