## Credit Card Fraud Detection using Decision Tree

```python
#import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import keras

np.random.seed(2)
```

```python
data = pd.read_csv('data/creditcard.csv')
data.head()
```

Here the data seems to be only numbers, because we can not have a origincal datas with the credit card information.

```python
#data preprocessing
from sklearn.preprocessing import StandardScaler
data['normalizedAmount'] = StandardScaler().fit_transform(data['Amount'].values.reshape(-1,1))
data = data.drop('Amount', axis=1)
data.head()
```

```python
#drop time column as it is not contibuting in predicting the fraudalent transaction
data = data.drop('Time', axis=1)
data.head()
```

```python
#Define X and y for the model input
X = data.iloc[:, data.columns != 'Class']
y = data.iloc[:, data.columns == 'Class']
```

```python
X.head()
```

```python
y.head()
```

```python
#split the data into training and testing dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

## Decision Tree

```python
#fit the Decision Tree model into the training dataset
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train)
```

```python
#predict the results for the testing dataset
y_pred = classifier.predict(X_test)
```

```python
#calculate the score for the testing dataset
classifier.score(X_test, y_test)
```

```python
#confusion matrix for the testing dataset
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(10,5))
sns.set(font_scale=1.4)
sns.heatmap(cm, annot=True)
plt.ylabel('Predicted', fontsize=15)
plt.xlabel('True', fontsize=15)
plt.title('Confusion matrix', fontsize=15)
cm
```

```python
#predict the results for the entire dataset
y_pred_all = classifier.predict(X)
```

```python
#calculate the score for the entire dataset
classifier.score(X, y)
```

```python
#confusion matrix for the entire dataset
from sklearn.metrics import confusion_matrix
cm_all = confusion_matrix(y, y_pred_all)
plt.figure(figsize=(10,5))
sns.set(font_scale=1.4)
sns.heatmap(cm_all, annot=True)
plt.ylabel('Predicted', fontsize=15)
plt.xlabel('True', fontsize=15)
plt.title('Confusion matrix', fontsize=15)
cm_all
```
